```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


# Define the URL to download
url = "http://bigg.ucsd.edu/api/v2/models/Recon3D/reactions"

# Send a GET request to the URL and get the response
response = requests.get(url)

# Decode the response content from binary to text
content = response.content.decode("utf-8")

# Convert the response to a pandas DataFrame
df = pd.read_json(content)

# Extract the values from the dictionaries in the first column of the DataFrame
reactions = pd.json_normalize(df.iloc[:, 0])

bigg_ids = reactions.iloc[:, 0]

combined_df = []
i = 0
#Download each file and append contents to a single CSV file
with open('reactions.csv', 'w') as f:
    for bigg_id in bigg_ids:
        print(['downloading ', i])
        i = i + 1
        download_url = f'http://bigg.ucsd.edu/models/Recon3D/reactions/{bigg_id}'
        response = requests.get(download_url)
        # Create a BeautifulSoup object from the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # extract the information you want
        descriptive_name = soup.find('h4', text='Descriptive name: ').find_next('p').text
        reaction = soup.find('h4', text='Reaction:').find_next('p').text
        subsystem = soup.find('h4', text='Subsystem:').find_next('p').text
        default_bounds = soup.find('h4', text='Default bounds:').find_next('p').text
        gene_reaction_rule = soup.find('h4', text='Gene Reaction Rule:').find_next('p').text
        genes = soup.find('h4', text='Genes:').find_next('p').text

        # create a Pandas DataFrame
        df = pd.DataFrame({
            'bigg_id': [bigg_id],
            'descriptive_name': [descriptive_name],
            'reaction': [reaction],
            'subsystem': [subsystem],
            'default_bounds': [default_bounds],
            'gene_reaction_rule': [gene_reaction_rule],
            'genes': [genes]
        })
        # Add the dataframe to the list
        combined_df.append(df)

# Read CSV file into DataFrame
final_df = pd.concat(combined_df, ignore_index=True)
final_df.to_csv('extdata/RECON3/reactions_metadata.csv', index=False)
```
